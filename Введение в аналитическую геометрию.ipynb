2. Задание (а листочке)
Почему прямые не кажутся перпендикулярными? (см.ролик)

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y1 = (-1/3)*x+1
plt.plot(x,y)
plt.plot(x,y1)

# Прямые не перпендикулярны из-за непропорциональности графика по вертикали и по горизонтали
```

Исправим

```python
x = np.linspace(-5, 5, 21)
y = 3*x+1
y1 = (-1/3)*x+1

plt.plot(x,y)
plt.plot(x,y1)

plt.gca().set_aspect('equal')
plt.show()
```

```python
import matplotlib.pyplot as plt
circle = plt.Circle((0.5, 0.5), 0.5)
fig, ax = plt.subplots()
plt.grid(color='lightgray',linestyle='--')
ax.add_artist(circle)
plt.gca().set_aspect('equal')
plt.show()
```

Элипсы

```python
y = np.linspace(-5, 5, 2000)
a=20
b=5
x = np.sqrt((a**2*(1-y**2/b**2)))
x2 = -np.sqrt((a**2*(1-y**2/b**2)))
plt.plot(x,y)
plt.plot(x2,y)
plt.gca().set_aspect('equal')
plt.show()
```

Гипербола

```python
x = np.linspace(-10, 10, 50)
a=2
b=2

y = np.sqrt((b**2*(1+x**2/a**2)))
y2 = -np.sqrt((b**2*(1++x**2/a**2)))

plt.plot(x,y)
plt.plot(x,y2)

plt.show()
```

5. Задание (в программе)
Нарисуйте трехмерный график двух параллельных плоскостей.
Нарисуйте трехмерный график двух любых поверхностей второго порядка.


```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 5)
Y = np.arange(-5, 5, 5)
X, Y = np.meshgrid(X, Y)
Z = X
Z2 = X+2
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```

```python
from pylab import *
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = Axes3D(fig)
X = np.arange(-5, 5, 0.3)
Y = np.arange(-5, 5, 0.3)
X, Y = np.meshgrid(X, Y)
Z = np.sqrt(X*X + Y*Y)
Z2 = X*X - Y*Y
ax.plot_surface(X, Y, Z)
ax.plot_surface(X, Y, Z2)
ax.scatter(0, 0, 0,'z',50,'red')
show()
```
