3.1. (Сделал как понял): Постройте на одном графике две кривые y(x) для функции двух переменной y(k,x)=cos(k∙x), взяв для одной кривой значение k=1, а для другой – любое другое k, не равное 1.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0, 10, 100)
plt.plot(x, np.cos(x))
k = 1
plt.plot(x, np.cos(x*k))
k = 1.2
plt.plot(x, np.cos(x*k))
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
